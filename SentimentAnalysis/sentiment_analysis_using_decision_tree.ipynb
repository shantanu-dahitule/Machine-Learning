```
cols=['sentiment','id','date','query_string','user','text']
data=pd.read_csv('/content/drive/MyDrive/NOTES BACKUP/DATAsets/training_1600000_processed_noemoticon.csv',encoding='latin1', names=cols)
```

```
data.drop(['id','date','query_string','user'],axis=1,inplace=True)
```

```
data
```

```
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```
def format_text(df,col):
  #Remove @ tags
  comp_df = df.copy()
    
  # remove all the punctuation
  comp_df[col] = comp_df[col].str.replace(r'(@\w*)','')

  #Remove URL
  comp_df[col] = comp_df[col].str.replace(r"http\S+", "")

  #Remove # tag and the following words
  comp_df[col] = comp_df[col].str.replace(r'#\w+',"")

  #Remove all non-character
  comp_df[col] = comp_df[col].str.replace(r"[^a-zA-Z ]","")

  # Remove extra space
  comp_df[col] = comp_df[col].str.replace(r'( +)'," ")
  comp_df[col] = comp_df[col].str.strip()

  # Change to lowercase
  comp_df[col] = comp_df[col].str.lower()

  return comp_df

formated_df = format_text(data,'text')
```

```
X_data=formated_df.drop('sentiment',axis=1)
Y_data=pd.DataFrame(formated_df['sentiment'])
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_data,Y_data)
```

```
y_train
```

```
# tokenizing and stop word removal
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
def tknize_and_stop(tweet):
  tok_li=[]
  stop=set(stopwords.words('english'))
  for i in tweet:
    li=word_tokenize(i)
    stop_rem=[i for i in li if i not in stop]
    sen=" ".join(stop_rem)
    tok_li.append(sen)
  
  return tok_li
```

```
# Stemmer
from nltk.stem import SnowballStemmer
def stem(tweet):
  ss=SnowballStemmer('english')
  stem_sen=[]
  for i in tweet:
    sentence=''
    wrds=i.split(' ')
    for i in wrds:
      w=ss.stem(i)
      w=w.lower()
      sentence+=w
      sentence+=' '
    stem_sen.append(sentence)  
  return stem_sen
```

```
X_li=list(X_train['text'])
```

```
tkn_li=tknize_and_stop(X_li)
```

```
tkn_li
```

```
X_data_stem=stem(tkn_li)
```

```
X_data_stem
```

```
from sklearn.feature_extraction.text import CountVectorizer
c_vec=CountVectorizer(max_features=20000)
vectorized=c_vec.fit(X_data_stem)
```

```
vectorized=c_vec.transform(X_data_stem)
```

```
#using decision tree
from sklearn.tree import DecisionTreeClassifier
```

```
clf=DecisionTreeClassifier(max_features='sqrt')
clf.fit(vectorized,y_train.values.ravel())
```

```
from sklearn.model_selection import cross_val_score 
print(cross_val_score(clf,vectorized,y_train.values.ravel(),cv=3,scoring='accuracy'))
```

```


# Transforming the test data
test=list(X_test['text'])
test_tkn=tknize_and_stop(test)
test_stem=stem(test_tkn)
```

```
test_vec=c_vec.transform(test_stem)
```

```
y_pred=clf.predict(test_vec)
```

```
from sklearn.metrics import f1_score,precision_score,recall_score
prec=precision_score(y_test.values.ravel(),y_pred,pos_label=4)
rec=recall_score(y_test.values.ravel(),y_pred,pos_label=4)
f1=f1_score(y_test.values.ravel(),y_pred,pos_label=4)
print('Precision', prec)
print('---------------------')
print('Recall', rec)
print('-------------------------')
print('F1 score', f1)




```
