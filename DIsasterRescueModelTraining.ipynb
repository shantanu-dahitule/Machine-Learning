```
import numpy as np
import pandas as pd
import sqlite3
import pickle

database = sqlite3.connect('/content/DisasterResponse.db')
query = database.execute("SELECT * From Message;")
cols = [column[0] for column in query.description]
data = pd.DataFrame.from_records(data = query.fetchall(), columns = cols)
```

```
X = data.iloc[:,2]
Y = data.iloc[:, 5:]
category_names = list(data.columns[5:])
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```
import nltk
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer 

nltk.download('punkt')

X_train = [a.lower() for a in X_train]

X_train = [word_tokenize(a) for a in X_train]

nltk.download('wordnet')


for i, sent in enumerate(X_train):
  for j, word in enumerate(sent):
    X_train[i][j] = WordNetLemmatizer().lemmatize(word, pos= 'n')


from nltk.corpus import stopwords 
nltk.download('stopwords')
stop_words = set(stopwords.words('english')) 


filtered_sent = []

for i, sent in enumerate(X_train):
  filtered_sent.append([word for word in sent if not word in stop_words])



new = []

for i in filtered_sent:
  listToStr = ' '.join(map(str, i)) 
  new.append(listToStr)



for i, sent in enumerate(filtered_sent):
  for j, word in enumerate(sent):
    X_train[i][j] = WordNetLemmatizer().lemmatize(word, pos='v')

sent = []
for i in X_train:
  message = ' '.join(map(str, i))
  sent.append(message)



import string
punctuation = set(string.punctuation)
no_punct = []
for top, i in enumerate(sent):
  a = list(i)
  # print(a)
  for index, k in enumerate(a):
    if k in punctuation:
      a[index] = ""
  b = ''.join(a)
  no_punct.append(b)


import nltk

from nltk.stem.lancaster import LancasterStemmer

stemmer = LancasterStemmer()

stemmed = []
for i in no_punct:
  a = i.split(" ")
  for j in a:
    j = stemmer.stem(j)
  # print(a)


lem_sentences = []
for i in no_punct:
  sentence = ''
  ori = i.split(' ')
  for j in ori:
    word = stemmer.stem(j)
    word = word.lower()
    sentence += word
    sentence += ' '
  lem_sentences.append(sentence)

from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

vectorizer = CountVectorizer()
tfidf = TfidfTransformer()

vectorizer.fit(lem_sentences)


vector = vectorizer.transform(lem_sentences)
tfidf.fit(vector)
tfidf_vector_train = tfidf.transform(vector)

print(vectorizer.vocabulary_)
print(vector.shape)
print(tfidf_vector_train.shape)
```

```
import nltk
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer 

nltk.download('punkt')

X_test = [a.lower() for a in X_test]

X_test = [word_tokenize(a) for a in X_test]

nltk.download('wordnet')


for i, sent in enumerate(X_test):
  for j, word in enumerate(sent):
    X_test[i][j] = WordNetLemmatizer().lemmatize(word, pos= 'n')


from nltk.corpus import stopwords 
nltk.download('stopwords')
stop_words = set(stopwords.words('english')) 


filtered_sent = []

for i, sent in enumerate(X_test):
  filtered_sent.append([word for word in sent if not word in stop_words])



new = []

for i in filtered_sent:
  listToStr = ' '.join(map(str, i)) 
  new.append(listToStr)



for i, sent in enumerate(filtered_sent):
  for j, word in enumerate(sent):
    X_test[i][j] = WordNetLemmatizer().lemmatize(word, pos='v')

sent = []
for i in X_test:
  message = ' '.join(map(str, i))
  sent.append(message)



import string
punctuation = set(string.punctuation)
no_punct = []
for top, i in enumerate(sent):
  a = list(i)
  # print(a)
  for index, k in enumerate(a):
    if k in punctuation:
      a[index] = ""
  b = ''.join(a)
  no_punct.append(b)


import nltk

from nltk.stem.lancaster import LancasterStemmer

stemmer = LancasterStemmer()

stemmed = []
for i in no_punct:
  a = i.split(" ")
  for j in a:
    j = stemmer.stem(j)
  # print(a)


lem_sentences = []
for i in no_punct:
  sentence = ''
  ori = i.split(' ')
  for j in ori:
    word = stemmer.stem(j)
    word = word.lower()
    sentence += word
    sentence += ' '
  lem_sentences.append(sentence)

from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

# vectorizer = CountVectorizer()
# tfidf = TfidfTransformer()

# vectorizer.fit(lem_sentences)


vector_test = vectorizer.transform(lem_sentences)
# tfidf.fit(vector)
tfidf_vector_test = tfidf.transform(vector_test)

# print(vectorizer.vocabulary_)
print(vector_test.shape)
print(tfidf_vector_test.shape)
```

```
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier



model = MultiOutputClassifier(RandomForestClassifier())




model.fit(tfidf_vector_train, Y_train)


from sklearn.metrics import classification_report, accuracy_score



Y_pred = model.predict(tfidf_vector_test.toarray())
acc = 0
for i in range(len(category_names)):
    print("Category:", category_names[i],"\n", classification_report(Y_test.iloc[:, i].values, Y_pred[:, i]))
    print('Accuracy of %25s: %.2f' %(category_names[i], accuracy_score(Y_test.iloc[:, i].values, Y_pred[:,i])))
    acc += accuracy_score(Y_test.iloc[:, i], Y_pred[:,i])

acc / len(category_names)


pickle.dump(model, open("/content/drive/My Drive/Models/RandomForest.pkl", "wb"))
```
